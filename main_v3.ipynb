```python
# main.ipynb

import numpy as np
from parse_script import parse_marker_data
from calculations import compute_quantities
from detection import detect_collision_times
from visualization import plot_trajectories, plot_quantity_over_time, compare_before_after


# Load and parse data
filepath = 'data/Dag2/1De_v75-7g_h62-1g_4.tsv'
markers_dict, time, masses = parse_marker_data(filepath)
results = compute_quantities(markers_dict, time, masses)

# Plot the trajectories
plot_trajectories(markers_dict, time)

# Detect collision times based on velocity changes
collision_times = detect_collision_times(markers_dict, time, threshold=0.1)
print("Detected Collision Times:", collision_times)

# Plot momentum and kinetic energy over time
plot_quantity_over_time(results, time, "momentum")
plot_quantity_over_time(results, time, "kinetic_energy")

# Analyze conservation around each detected collision
for collision_start, collision_end in collision_times:
    print(f"\nAnalyzing collision from {collision_start} to {collision_end}")
    momentum_diff, kinetic_energy_diff = compare_before_after(results, time, collision_start, collision_end)
```
